```sql
USE CATALOG catalogo;
USE SCHEMA ventas;


SELECT 

Categoria_Producto
,SUM(Total_Venta) AS Ventas_Totales
FROM catalogo.ventas.raw_ventas
GROUP BY Categoria_Producto;

--View dinamica
create or replace view Vista_Dinamica_Resumen as
SELECT 

Categoria_Producto
,SUM(Total_Venta) AS Ventas_Totales
FROM catalogo.ventas.raw_ventas
GROUP BY Categoria_Producto;




--View Materialized
create materialized view Vista_Materilizada_Resumen as
SELECT 
Categoria_Producto
,SUM(Total_Venta) AS Ventas_Totales
FROM catalogo.ventas.raw_ventas
GROUP BY Categoria_Producto;

--dinamica

SELECT * FROM Vista_Dinamica_Resumen;

--Materializada

select * from Vista_Materilizada_Resumen;


--Insertamos una nueva fila 
insert into catalogo.ventas.raw_ventas(
ID_Venta
,Fecha_Venta
,Categoria_Producto
,Total_Venta
,Region
,Canal_Venta
)
values('',current_date(),'Electronica',1000000,'Norte','Online');

refresh materialized view vista_Materilizada_Resumen;

DELETE FROM catalogo.ventas.raw_ventas WHERE ID_Venta = '';


--Vista Materializada
--select * from Vista_Materilizada_Resumen;

--Vista Dinamica
--select * from Vista_Dinamica_Resumen;
```
