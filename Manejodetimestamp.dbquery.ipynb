```sql
USE CATALOG catalogo;
USE SCHEMA ventas;


CREATE TABLE IF NOT EXISTS shipping_logs (
 ID_Envio STRING
 ,ID_Venta_Ref STRING
 ,Fecha_Envio_Str STRING
 );


 COPY INTO shipping_logs
 FROM '/Volumes/catalogo/filestore/tables/shoppin_logs.csv'
 FILEFORMAT = CSV
 FORMAT_OPTIONS ('header' = 'true');




SELECT


ID_Venta_Ref
,Fecha_Envio_Str
,to_date(Fecha_Envio_Str,'dd/MM/yyyy') AS Fecha_Limpia

FROM catalogo.ventas.shipping_logs;




SELECT
A.ID_Venta
,A.Fecha_Venta
,B.Fecha_Envio_Str
,to_date(Fecha_Envio_Str,'dd/MM/yyyy') AS Fecha_Limpia
,date_diff(to_date(Fecha_Envio_Str,'dd/MM/yyyy'),A.Fecha_Venta) AS Dias_Diferencia
FROM catalogo.ventas.raw_ventas AS A
JOIN catalogo.ventas.shipping_logs AS B
ON ID_Venta = ID_Venta_Ref;




WITH Ventas_Con_Envio as (
SELECT
A.ID_Venta
,A.Fecha_Venta
,B.Fecha_Envio_Str
,to_date(Fecha_Envio_Str,'dd/MM/yyyy') AS Fecha_Limpia
,date_diff(to_date(Fecha_Envio_Str,'dd/MM/yyyy'),A.Fecha_Venta) AS Dias_Diferencia
FROM catalogo.ventas.raw_ventas AS A
JOIN catalogo.ventas.shipping_logs AS B
ON ID_Venta = ID_Venta_Ref

)

SELECT 
date_trunc('week',fecha_venta) as Semana_Venta
,avg(Dias_Diferencia) as Proemdio_Dias_Envio 
from Ventas_Con_Envio
GROUP BY Semana_Venta
ORDER BY Semana_Venta ASC;
```
