```sql
--Arquitectura Medallon


USE CATALOG catalogo;


USE SCHEMA ventas;


--select * from catalogo.ventas.raw_ventas v;


select * from raw_ventas v;


CREATE OR REPLACE TABLE ventas_silver AS
SELECT 
 CASE WHEN Fecha_Venta IS NULL THEN split(ID_Venta,',')[0] ELSE ID_Venta END as ID_Venta
,try_cast(CASE WHEN Fecha_Venta IS NULL THEN split(ID_Venta,',')[1] ELSE Fecha_Venta END AS DATE)  as Fecha_Venta
,CASE WHEN Fecha_Venta IS NULL THEN split(ID_Venta,',')[2] ELSE ID_Producto END as ID_Producto
,CASE WHEN Fecha_Venta IS NULL THEN split(ID_Venta,',')[3] ELSE Nombre_Producto END as Nombre_Producto
,CASE WHEN Fecha_Venta IS NULL THEN split(ID_Venta,',')[4] ELSE Categoria_Producto END as Categoria_Producto
,try_cast(CASE WHEN Fecha_Venta IS NULL THEN split(ID_Venta,',')[5] ELSE Cantidad END AS BIGINT) as Cantidad
,CASE WHEN Fecha_Venta IS NULL THEN split(ID_Venta,',')[6] ELSE Precio_Unitario END as Precio_Unitario
,try_cast(CASE WHEN Fecha_Venta IS NULL THEN split(ID_Venta,',')[7] ELSE Total_Venta END as double) as Total_Venta
,CASE WHEN Fecha_Venta IS NULL THEN split(ID_Venta,',')[8] ELSE Region END as Region
,CASE WHEN Fecha_Venta IS NULL THEN split(ID_Venta,',')[9] ELSE Canal_Venta END as Canal_Venta



FROM raw_ventas v;
```
