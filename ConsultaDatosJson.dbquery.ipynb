```sql
-- Ejemplo 1: Crear un JSON como VARIANT y mostrarlo completo

WITH Tabla_Usuarios_JSON AS (
    SELECT
        parse_json('{
            "ID_Usuario": "U-001",
            "Nombre": "Ana García",
            "Email": "ana.garcia@email.com",
            "Regiones_Interes": ["Norte", "Sur", "Este"],
            "Direccion": {
                "Calle": "Calle Falsa 123",
                "Ciudad": "Sevilla",
                "Pais": "España"
            }
        }') AS datos_usuario
)

SELECT datos_usuario
FROM Tabla_Usuarios_JSON;



-- Ejemplo 2: Acceder a propiedades simples y objeto completo

WITH Tabla_Usuarios_JSON AS (
    SELECT
        parse_json('{
            "ID_Usuario": "U-001",
            "Nombre": "Ana García",
            "Email": "ana.garcia@email.com",
            "Regiones_Interes": ["Norte", "Sur", "Este"],
            "Direccion": {
                "Calle": "Calle Falsa 123",
                "Ciudad": "Sevilla",
                "Pais": "España"
            }
        }') AS datos_usuario
)

SELECT
    datos_usuario:Email      AS Email_Usuario,
    datos_usuario:Direccion  AS Direccion_Completa,
    datos_usuario:Nombre  AS Nombre
FROM Tabla_Usuarios_JSON;


-- Ejemplo 3: Acceder a campos dentro del objeto Direccion

WITH Tabla_Usuarios_JSON AS (
    SELECT
        parse_json('{
            "ID_Usuario": "U-001",
            "Nombre": "Ana García",
            "Email": "ana.garcia@email.com",
            "Regiones_Interes": ["Norte", "Sur", "Este"],
            "Direccion": {
                "Calle": "Calle Falsa 123",
                "Ciudad": "Sevilla",
                "Pais": "España"
            }
        }') AS datos_usuario
)

SELECT
    datos_usuario:Direccion.Ciudad AS Ciudad_Residencia,
    datos_usuario:Direccion.Pais   AS Pais_Residencia
FROM Tabla_Usuarios_JSON;


-- Ejemplo 4: Acceder a un array completo y por posición

WITH Tabla_Usuarios_JSON AS (
    SELECT
        parse_json('{
            "ID_Usuario": "U-001",
            "Nombre": "Ana García",
            "Email": "ana.garcia@email.com",
            "Regiones_Interes": ["Norte", "Sur", "Este"],
            "Direccion": {
                "Calle": "Calle Falsa 123",
                "Ciudad": "Sevilla",
                "Pais": "España"
            }
        }') AS datos_usuario
)

SELECT
    datos_usuario:Regiones_Interes      AS Array_Completo,
    datos_usuario:Regiones_Interes[0]   AS Primera_Region,
    datos_usuario:Regiones_Interes[1]   AS Segunda_Region,
    datos_usuario:Regiones_Interes[2]   AS Tercera_Region
FROM Tabla_Usuarios_JSON;




-- Ejemplo 5 corregido: Convertir el array VARIANT en ARRAY tipado antes de explode

WITH Tabla_Usuarios_JSON AS (
    SELECT
        parse_json('{
            "ID_Usuario": "U-001",
            "Nombre": "Ana García",
            "Email": "ana.garcia@email.com",
            "Regiones_Interes": ["Norte", "Sur", "Este"],
            "Direccion": {
                "Calle": "Calle Falsa 123",
                "Ciudad": "Sevilla",
                "Pais": "España"
            }
        }') AS datos_usuario
)
SELECT
    explode(cast(datos_usuario:Regiones_Interes AS ARRAY<STRING>)) AS region
FROM Tabla_Usuarios_JSON;
```
